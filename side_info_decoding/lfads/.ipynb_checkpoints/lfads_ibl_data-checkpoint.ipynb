## Import necessary modules

```python
%matplotlib inline

import torch
import torchvision
# np = torch._np
import numpy as np
import matplotlib.pyplot as plt

import os
import yaml

from lfads import LFADS_Net
from utils import read_data, load_parameters, save_parameters

from side_info_decoding.utils import set_seed, load_data_from_pids

seed = 666
set_seed(seed)

# plt.style.use('dark_background')
```

```python
# Select device to train LFADS on
device = 'cuda' if torch.cuda.is_available() else 'cpu'; print('Using device: %s'%device)
# device = 'cpu'
```

## Load or Generate Data 

```python
pids = [
    # "dab512bd-a02d-4c1f-8dbc-9155a163efc0",
    "31f3e083-a324-4b88-b0a4-7788ec37b191",
    # "84bb830f-b9ff-4e6b-9296-f458fb41d160",
]
```

```python
X_dict, Y_dict = load_data_from_pids(
    pids,
    brain_region="all",
    behavior="wheel_speed",
    data_type="good_ks",
    n_t_bins = 60
)
```

```python
pid = pids[0]
X, Y = X_dict[pid], Y_dict[pid]
X = X.transpose(0,2,1)
```

```python
from sklearn.model_selection import train_test_split, KFold

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=seed)
```

```python
all_data = torch.Tensor(X).to(device)
train_data = torch.Tensor(X_train).to(device)
valid_data = torch.Tensor(X_test).to(device)

train_ds      = torch.utils.data.TensorDataset(train_data)
valid_ds      = torch.utils.data.TensorDataset(valid_data)

num_trials, num_steps, num_cells = X.shape
print(train_data.shape)
```

## Load model hyperparameters 

```python
hyperparams = {'betas': (0.9, 0.99),
 # 'c_encoder_dim': 128//6,
 # 'clip_val': 15.0,
 # 'controller_dim': 128//6,
 'dataset_name': 'chaotic_rnn',
 # 'epsilon': 0.1,
 # 'factors_dim': 20,
 # 'g0_encoder_dim': 200//6,
 # 'g0_prior_kappa': 0.1,
 # 'g_dim': 200//6,
 # 'keep_prob': 0.95,
 # 'kl_weight_schedule_dur': 2000,
 # 'kl_weight_schedule_start': 0,
 # 'l2_con_scale': 0,
 # 'l2_gen_scale': 2000,
 # 'l2_weight_schedule_dur': 2000,
 # 'l2_weight_schedule_start': 0,
 'learning_rate': 0.0005,
 # 'learning_rate_decay': 0.95,
 # 'learning_rate_min': 1e-05,
 # 'max_norm': 200,
 # 'run_name': 'demo',
 # 'scheduler_cooldown': 6,
 # 'scheduler_on': True,
 # 'scheduler_patience': 6,
 # 'u_dim': 1,
 # 'u_prior_kappa': 0.1
}

hyperparams
```

## Instantiate LFADS model

```python
model = LFADS_Net(inputs_dim = num_cells, T = num_steps, dt = 0.025, device=device,
                 model_hyperparams=hyperparams).to(device)
```

## Fit model

Rule of thumb: You can usually see good fit after 200 epochs (~30 mins runtime on Thinkpad GPU, ~2.5 hours on CPU), but to see good inference of perturbation timings need to run for about 800 epochs (~2 hours on Thinkpad GPU).

```python
model.fit(train_ds, valid_ds, max_epochs=200, batch_size=32, use_tensorboard=False,
          train_truth=None, valid_truth=None)
```

### Inference

```python
recon_rates = np.zeros((num_trials, num_steps, num_cells))
for i in range(len(all_data)):
    recon_rates[i] = model.reconstruct(all_data[i][None,:])
```

```python
plt.figure(figsize = (12,12))
plt.imshow(recon_rates.mean(0).T/0.05, aspect="auto")
plt.xticks(np.linspace(0, 30, 6), ['%.1f'%i for i in np.linspace(0, 1.5, 6)])
plt.xlabel('Time (s)')
plt.ylabel('Cell #')
plt.colorbar(orientation='horizontal', label='Firing Rate (Hz)')
plt.title('Example trial');
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import explained_variance_score, r2_score
from scipy.stats import pearsonr
from sklearn.preprocessing import normalize


kf = KFold(n_splits=5, shuffle=True, random_state=seed)

X = recon_rates.reshape(-1, recon_rates.shape[1] * recon_rates.shape[2])
Y = normalize(Y)

obs_lst, pred_lst, r2_lst, corr_lst = [], [], [], []
for fold_idx, (train, test) in enumerate(kf.split(X)):
    train_X, test_X = X[train], X[test]
    train_Y, test_Y = Y[train], Y[test]
    lr = Ridge(alpha=1000).fit(train_X, train_Y,)
    pred_Y = lr.predict(test_X)
    obs_lst.append(test_Y)
    pred_lst.append(pred_Y)
    r2_lst.append(np.mean([r2_score(test_Y[k], pred_Y[k]) for k in range(len(test_Y))]))
    corr_lst.append(np.mean([pearsonr(test_Y[k], pred_Y[k])[0] for k in range(len(test_Y))]))

obs_lst = np.vstack(obs_lst)     
pred_lst = np.vstack(pred_lst)    
print(f"5-fold CV r2: {np.mean(r2_lst)} corr: {np.mean(corr_lst)}")
```
