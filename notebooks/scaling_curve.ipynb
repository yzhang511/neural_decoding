```python
import matplotlib.pyplot as plt
import numpy as np

# Data points for x-axis (number of sessions)
sessions = [1, 5, 10, 20, 40, 73]

# Data points for each metric (extracted from the graph)
choice = [0.82326, 0.90220, 0.90257, 0.91515, 0.90863, 0.90890]
prior = [0.69830, 0.70828, 0.71209, 0.72685, 0.73131, 0.70599]
wheel = [0.22170, 0.22041, 0.24144, 0.25912, 0.25042, 0.25002]
whisker = [0.34130, 0.35890, 0.36134, 0.37067, 0.37220, 0.37648]
```

```python
# Create figure and axes
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
plt.subplots_adjust(wspace=0.3)

title_font = {'fontsize': 22}
tick_font = {'fontsize': 18}

metrics = [
    (choice, "Choice", "Accuracy"),
    (prior, "Prior", "Correlation"),
    (wheel, "Wheel", r"$R^2$"),
    (whisker, "Whisker", r"$R^2$")
]

for i, (data, title, ylabel) in enumerate(metrics):
    axes[i].plot(sessions, data, 'o-', color='dodgerblue', linewidth=3, markersize=10)
    axes[i].set_title(title, pad=10, **title_font)
    axes[i].set_xlabel("# Sessions", fontsize=22)
    axes[i].set_ylabel(ylabel, fontsize=22)
    axes[i].grid(True, linestyle='--', alpha=0.7, linewidth=2., axis='y')
    axes[i].set_xticks(sessions)

    axes[i].tick_params(axis='both', which='major', labelsize=tick_font['fontsize'])

    # Set reasonable y-axis limits based on data range
    ymin, ymax = min(data), max(data)
    margin = (ymax - ymin) * 0.1
    y_range = [ymin - margin, ymax + margin]
    axes[i].set_ylim(y_range)

    axes[i].set_xlim(-5, 80)

    # Create 5 evenly spaced ticks within the y-range
    yticks = np.linspace(y_range[0], y_range[1], 5)
    axes[i].set_yticks(yticks)

    axes[i].yaxis.set_major_formatter(plt.FormatStrFormatter('%.2f'))

    for spine in axes[i].spines.values():
      spine.set_linewidth(2)

    axes[i].tick_params(axis='both', which='major', labelsize=tick_font['fontsize'], width=2, length=6)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Save the figure
plt.savefig('neuron_revision_scaling.png', dpi=400, bbox_inches='tight')
plt.show()
plt.close()
```
